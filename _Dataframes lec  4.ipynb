# Replacing the missing value in dataframe
# mode means which value is repeated more time in columns,rows
# 1. Categorical columns [means the text data] = Replace by Mode
# 2. Continuous values[ Numerical data] = Replace by Mean , Median
# 3. function used to replace missing values is df.fillna(value)

```python
df["AirBags"].value_counts()
```

```python
md = df["AirBags"].mode()[0]
```

# mode is applicable when limited missing values are available
# More value are missing then drop the columns

```python
df['AirBags'] = df['AirBags'].fillna(md)
```

```python
df.isna().sum()
```

```python
df
```

```python
g = df["Rear.seat.room"].mean().round(2)
```

```python
g
```

```python
df['Rear.seat.room'] = df['Rear.seat.room'].fillna(g)
```

```python
df.isna().sum()
```

```python
median = df['Luggage.room'].median()
```

```python
median
```

```python
df['Luggage.room'] = df['Luggage.room'].fillna(median)
```

# Create a column based on condition

```python
def price_category(price):
    if price < 17:
        price_cat = "Cheap"
    elif price >= 17 and price <= 22:
        price_cat = "Moderate"
    else :
        price_cat = "Expensive"
    return price_cat
```

```python
 price_cat(34)
```

```python
price_cat(4)
```

```python
price_cat(20)
```

```python
df["Price_Category"] = df["Price"].apply(price_category)
```

```python
df
```

```python
df["Price_Category"].value_counts()
```

```python
df.groupby(by ="Price_Category").agg({"Price" : "mean"}).round(2)
```
